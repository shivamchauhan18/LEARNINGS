#### Assignment-13 : Functions & Flexible Arguments

```python
# 1. Function with Parameters

def add(a, b):
    return a + b

print(add(10, 20))
```

```python
# # 2. Default Parameters

def greet(name, message="Hello"):
    print(f"{message}, {name}")

greet("Shivam")
greet("Rahul", "Good Morning")
```

```python
# Keyword Argument
def details(name, age, city):
    print("Name:", name)
    print("Age:", age)
    print("City:", city)

details(name="Shivam", age=23, city="Hyderabad")

print()

details(city="Delhi", name="Rahul", age=25)
```

```python
# *args

def total(*args):
    return sum(args)

print(total(10, 20))
print(total(10, 20, 30, 40))
```

```python
# 5. **kwargs

def information(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")

information(name="Shivam", age=23, city="Hyderabad")
```

```python
def calculate_price(price, discount=0, *items, **kwargs):
    total = price

    for item in items:
        total += item

    total -= discount

    print("Total Price:", total)

    for key, value in kwargs.items():
        print(f"{key}: {value}")

calculate_price(
    1000,
    100,
    200,
    300,
    delivery=50,
    tax=18
)
```

```python
add = lambda x, y: x + y
square = lambda x: x ** 2

print(add(10, 20))
print(square(8))
```

```python
numbers = [1, 2, 3, 4, 5]

result = list(map(lambda x: x ** 2, numbers))

print(result)
```

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

even = list(filter(lambda x: x % 2 == 0, numbers))

print(even)
```
