#### Assignment-14: Exception Handling in Python

```python
# Generate an Exception

print(10/0)
```

```python
# Handle Execption
try:
    print(10 / 0)
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# 3. Multiple Except Blocks


try:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print(a / b)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Please enter numbers only")
```

```python
# 4. Raise Exception


age = int(input("Enter age: "))

if age < 18:
    raise Exception("Age must be 18 or above")

print("Eligible")
```

```python
# 5. Function That Raises Exception


def demo():
    raise Exception("This is an exception")

try:
    demo()

except Exception as e:
    print(e)
```

```python
class InsufficientBalanceError(Exception):
    pass

balance = 500
withdraw = 1000

try:
    if withdraw > balance:
        raise InsufficientBalanceError("Not enough balance")

except InsufficientBalanceError as e:
    print(e)

```

```python
try:
    file = open("sample.txt", "r")
    print(file.read())

except FileNotFoundError:
    print("File not found")

finally:
    print("Program Finished")
```

```python
# 8. File Not Found


try:
    file = open("abc.txt", "r")

except FileNotFoundError:
    print("File does not exist")
```

```python
# 9. Type Error


try:
    print(10 + "20")

except TypeError:
    print("Type Error")
```

```python
# 10. Attribute Error


class Student:
    pass

s = Student()

try:
    print(s.name)

except AttributeError:
    print("Attribute not found")
```

```python
# 11. Index Error

numbers = [10, 20, 30]

try:
    print(numbers[5])

except IndexError:
    print("Invalid Index")
```

```python
# 12. else Block

try:
    number = int(input("Enter a number: "))
    print(number)

except ValueError:
    print("Invalid Input")

else:
    print("No Exception")
```

```python
# 13. Logging Errors


import logging

logging.basicConfig(filename="error.log")

try:
    print(10 / 0)

except Exception as e:
    logging.error(e)
    print("Error Logged")
```

```python

# 14. Input Validation

while True:

    try:
        age = int(input("Enter age: "))
        break

    except ValueError:
        print("Please enter a valid number")

print("Age:", age)
```
