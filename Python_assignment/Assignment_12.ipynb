```python
# ==========================================
# Assignment-12 : Constructors in Python
# ==========================================

# ------------------------------------------
# 1. Default and Parameterized Constructors
# ------------------------------------------

print("Question 1")

class Student:
    def __init__(self, name="Unknown", age=0):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")


s1 = Student()
s2 = Student("Shivam")
s3 = Student("Rahul", 23)

s1.display()
s2.display()
s3.display()


# ------------------------------------------
# 2. Calling Parent Constructor
# ------------------------------------------

print("\nQuestion 2")

class Parent:
    def __init__(self):
        print("Parent Constructor")
        self.parent_name = "Parent"


class Child(Parent):
    def __init__(self):
        super().__init__()
        self.child_name = "Child"
        print("Child Constructor")


c = Child()

print(c.parent_name)
print(c.child_name)


# ------------------------------------------
# 3. Simulating Access Levels
# ------------------------------------------

print("\nQuestion 3")

class Demo:
    def __init__(self):
        self.public = "Public Variable"
        self._protected = "Protected Variable"
        self.__private = "Private Variable"

    def show_private(self):
        print(self.__private)


d = Demo()

print(d.public)
print(d._protected)

# print(d.__private)   # Error

d.show_private()

print(d._Demo__private)     # Name Mangling


# ------------------------------------------
# 4. Constructor Attributes
# ------------------------------------------

print("\nQuestion 4")

class Employee:
    def __init__(self, name, emp_id, salary):
        self.name = name
        self.emp_id = emp_id
        self.salary = salary


e = Employee("Shivam", 101, 50000)

print(e.name)
print(e.emp_id)
print(e.salary)

print(e.__dict__)


# ------------------------------------------
# 5. __str__()
# ------------------------------------------

print("\nQuestion 5")

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}"


s = Student("Shivam", 23)

print(s)


# ------------------------------------------
# 6. Constructor with *args
# ------------------------------------------

print("\nQuestion 6")

class Numbers:
    def __init__(self, *args):
        self.values = args

    def display(self):
        print(self.values)


n = Numbers(10, 20, 30, 40, 50)

n.display()


# ------------------------------------------
# 7. Real World Example
# ------------------------------------------

print("\nQuestion 7")

class Employee:
    def __init__(self, name, emp_id, salary):
        self.name = name
        self.emp_id = emp_id
        self.salary = salary

    def display(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Salary: {self.salary}")
        print()


e1 = Employee("Shivam", 101, 50000)
e2 = Employee("Rahul", 102, 60000)
e3 = Employee("Amit", 103, 70000)

e1.display()
e2.display()
e3.display()
```
