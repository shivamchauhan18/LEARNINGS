#### Assignment-15: Dictionaries in Python

```python

# 1. Create a Dictionary


students = {
    101: "Shivam",
    102: "Rahul",
    103: "Amit",
    104: "Priya",
    105: "Neha"
}

print("Original Dictionary")
print(students)
```

```python
# 1.1 Add New Entries


students[106] = "Karan"

print("\nAfter Adding")
print(students)
```

```python

# 1.2 Update Existing Values

students[101] = "Shivam Chauhan"

print("\nAfter Updating")
print(students)
```

```python
# 1.3 Access Values

print("\nStudent with ID 103:", students[103])

print("\nAll Student Names")

for name in students.values():
    print(name)
```

```python
# 1.4 Iterate Through Dictionary
print("\nKeys and Values")

for key, value in students.items():
    print(key, ":", value)
```

```python
# 1.5 Print Only Keys
print("\nStudent IDs")

for key in students.keys():
    print(key)
```

```python
# 1.6 Print Only Values
print("\nStudent Names")

for value in students.values():
    print(value)
```

```python
# 1.7 Nested Dictionary

students = {
    101: {"name": "Shivam", "age": 23},
    102: {"name": "Rahul", "age": 22}
}

print("\nNested Dictionary")
print(students)
```

```python
# 1.8 Access Nested Values

print("\nStudent Name:", students[101]["name"])
print("Student Age:", students[101]["age"])
```

```python

# 1.9 Delete Elements
del students[102]
print(students)

students[103] = {"name": "Amit", "age": 24}
students.pop(103)
print(students)

students[104] = {"name": "Priya", "age": 21}

students.popitem()

print("\nAfter popitem()")
print(students)
```

```python
# 2. Check Key Existence
print("\nCheck Key")

student_id = 101

if student_id in students:
    print("Student ID Found")
else:
    print("Student ID Not Found")
```

```python
# 3. Count Entries

print("\nTotal Students")

print(len(students))
```

```python
# 4. Merge Two Dictionaries
dict1 = {
    101: "Shivam",
    102: "Rahul"
}

dict2 = {
    103: "Amit",
    104: "Neha"
}

merged = dict1 | dict2
merged
```

```python
# 5. Dictionary Comprehension

squares = {x: x*x for x in range(1,6)}

print("\nDictionary Comprehension")
print(squares)
```

```python
# 6. Reverse Dictionary

students = {
    101: "Shivam",
    102: "Rahul",
    103: "Amit"
}

reverse = {}

for key, value in students.items():
    reverse[value] = key

reverse    
```
