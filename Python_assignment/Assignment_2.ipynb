### Assignment-2: Operators in Python

```python
def maths(x, y):
    multi = x * y
    division = x / y
    addition = x + y
    subtraction = x - y

    return f"""Multiplication = {multi}
Division = {division}
Addition = {addition}
Subtraction = {subtraction}"""

print(maths(5, 5))
```

```python
# 2. Simulating Increment and Decrement
def increment_decrement(num):
    print("Original Value:", num)

    num += 1
    print("After Increment:", num)

    num -= 1
    print("After Decrement:", num)

increment_decrement(10)
```

```python
# 4. Relational Operators Demonstration
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("a < b :", a < b)
print("a <= b:", a <= b)
print("a > b :", a > b)
print("a >= b:", a >= b)
```

```python
# 5. Find Smaller and Larger Numbers

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a < b:
    print("Smaller Number:", a)
    print("Larger Number:", b)
elif b < a:
    print("Smaller Number:", b)
    print("Larger Number:", a)
else:
    print("Both numbers are equal.")     
```

```python
# 6. Combine Conditions (Largest of Three Numbers)
a = 5
b = 7
c = 12

if a >= b and a >= c:
    print("Largest Number:", a)
elif b >= a and b >= c:
    print("Largest Number:", b)
else:
    print("Largest Number:", c)


# Largest or smallest     
def small_large(a, b, c):
    large = a
    small = a

    if b > large:
        large = b
    if c > large:
        large = c

    if b < small:
        small = b
    if c < small:
        small = c

    return f"""Largest: {large}
Smallest: {small}"""

print(small_large(10, 11, 25))
```

```python
# 7. Operator-Based Calculator
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    print("Result:", a + b)
elif operator == "-":
    print("Result:", a - b)
elif operator == "*":
    print("Result:", a * b)
elif operator == "/":
    if b != 0:
        print("Result:", a / b)
    else:
        print("Division by zero is not allowed.")
else:
    print("Invalid operator.")
```
