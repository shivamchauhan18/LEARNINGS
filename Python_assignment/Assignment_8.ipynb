2. Validation Using Methods
Modify the Student class:
- Add validation:
- Age should be greater than 0

```python
class Student:
    def set_values(self, name, age):
        self.name = name

        if age > 0:
            self.age = age
        else:
            print("Invalid age! Age must be greater than 0.")
            self.age = None

    def get_values(self):
        print("Name:", self.name)
        print("Age:", self.age)


# Driver Code
student1 = Student()
student1.set_values("Shivam", 23)
student1.get_values()

print()

student2 = Student()
student2.set_values("Rahul", -5)
student2.get_values()
```

Rewrite the Student class using:

- @property → to get value
- @setter → to set value

```python
class Student:
    def __init__(self):
        self._name = ""
        self._age = 0

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        self._name = value

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if value > 0:
            self._age = value
        else:
            print("Age must be greater than 0.")


# Driver Code
student = Student()

student.name = "Shivam"
student.age = 23

print("Name:", student.name)
print("Age:", student.age)

student.age = -5
```

```python
# Read only property
class Employee:
    def __init__(self, emp_id):
        self._id = emp_id

    @property
    def id(self):
        return self._id


emp = Employee(101)

print("Employee ID:", emp.id)

# emp.id = 200     
```

```python
# Bank account system
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient Balance")

    def show_balance(self):
        print("Balance:", self.balance)


account = BankAccount("Shivam", 5000)

account.deposit(1000)
account.withdraw(2000)
account.withdraw(10000)

account.show_balance()
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    @property
    def area(self):
        return self.length * self.width


r = Rectangle(10, 5)

print("Area:", r.area)
```
