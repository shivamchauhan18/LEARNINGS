```python
# Write a function to calculate the sum of all elements in a list of integers.

def add_num(x):
    buffer = 0
    for i in x:
        buffer = i + buffer
    return buffer

x = [1,2,3,4]
add_num(x)
```

```python
# Write a function to identify duplicate values in a list.

x = [1,2,3,3,45,6,6,7]
x.sort()
for i in range(len(x)-1):
    if x[i] == x[i+1]:
        print(x[i])
```

```python
# Write a program to find common elements between two lists.

x= [1,2,3,4,5]
y= [3,5,8,9,7]

a = set(x)
b = set(y)

res = a.intersection(b)
print(res)
```

```python
# Write a program to count the frequency of each element in the list.

x = [1,2,2,3,3,4,5,6,6,6,6]

freq ={}

for i in x:
    if i in freq:
        freq[i] +=1
    else:
        freq[i] = 1 

print(freq)           
```

```python
# Write a program to count the frequency of each element in the list.

def sorting(x:list):
    if len(x) <1:
        return x
    pivot = x[0]

    left = []
    right = []

    for i in x[1:]:
        if i < pivot:
            left.append(i)
        else:
            right.append(i)

    return sorting(left) + [pivot] + sorting(right) 

x = [2,13,4,543,66,7,243]

sorting(x)

```
