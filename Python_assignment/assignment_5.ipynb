#### Assignment-5: Class Variables in Python

```python
# Access and print the variable using the class name
class Student:
    college = "Academy"   # Class variable

print(Student.college)
```

```python
# Access the same class variable using the instance
class Student:
    college = "Academy"

s1 = Student()    
s1.college
```

```python
# You modify the class variable using an instance

class Student:
    college = "Academy"

s1 = Student()    
s1.college = "MIT"
print(s1.college)
print(Student.college)
```

```python
# You modify the class variable using the class name
# Print the updated value using:
# - The class
# - Multiple instances

class Student:
    college = "Academy"

Student.college = "MIT"
Student.college    

s1 = Student()
s1.college
```

```python
# Create a class with:
# One class variable
# One instance variable

class Student:
    college = "Academy"

s2 = Student()
s2.college
s2.name = "shivam"

print(s2.college)
print(s2.name)
```
