#### Assignment-7: Inheritance & Polymorphism in Python

```python
class A:
    def __init__(self):
        self.value = "Value from A"

    def method1(self):
        print("Method 1 of A")

    def method2(self):
        print("Method 2 of A")

    def common_method(self):
        print("Common method from A")


class B(A):
    def __init__(self):
        super().__init__()
        self.value = "Value from B"

    def method3(self):
        print("Method 1 of B")

    def method4(self):
        print("Method 2 of B")

    def common_method(self):
        super().common_method()
        print("Common method from B")


class C(B):
    def __init__(self):
        super().__init__()
        self.value = "Value from C"

    def method5(self):
        print("Method 1 of C")

    def method6(self):
        print("Method 2 of C")

    def common_method(self):
        super().common_method()
        print("Common method from C")


# ---------------- Driver Code ----------------

a = A()
b = B()
c = C()

print("----- Class A -----")
a.method1()
a.method2()
a.common_method()

print("\n----- Class B -----")
b.method3()
b.method4()
b.common_method()

print("\n----- Class C -----")
c.method5()
c.method6()
c.common_method()

print("\n----- Runtime Polymorphism -----")

ref = B()
ref.common_method()

ref = C()
ref.common_method()

print("\n----- Instance Variables -----")

print(a.value)
print(b.value)
print(c.value)

ref = C()
print(ref.value)
```

```python
class Vehicle:
    def start_engine(self):
        print("Vehicle engine started")


class Car(Vehicle):
    def start_engine(self):
        print("Car engine started")


class ElectricCar(Car):
    def start_engine(self):
        print("Electric car is ready to drive")


v = Vehicle()
c = Car()
e = ElectricCar()

v.start_engine()
c.start_engine()
e.start_engine()
```
